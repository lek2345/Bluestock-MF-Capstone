```python
# AMFI Code Validation
import pandas as pd

fund_master = pd.read_csv("../data/raw/01_fund_master.csv")
nav = pd.read_csv("../data/raw/02_nav_history.csv")

fund_codes = set(fund_master["amfi_code"].unique())
nav_codes = set(nav["amfi_code"].unique())

missing_codes = fund_codes - nav_codes

print("Fund Master Codes:", len(fund_codes))
print("NAV History Codes:", len(nav_codes))
print("Missing Codes:", len(missing_codes))

if len(missing_codes) == 0:
    print("✅ All AMFI codes in fund_master exist in nav_history")
else:
    print("❌ Missing AMFI Codes:")
    print(missing_codes)
```
